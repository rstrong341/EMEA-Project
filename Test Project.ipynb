```python
def highlight_80(y):
    if y.Consumed > .8:
        return ['background-color: yellow']*14
    else:
        return ['background-color: white']*14
df.style.apply(highlight_80, axis=1)
```

```python
def equals(y):
    if y.EffectiveDate == 0.997050:
        return ['background-color: orange']*14
    else:
        return ['background-color: white']*14
df.style.apply(equals, axis=1)
```
